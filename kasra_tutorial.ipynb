### Libs

```python
import torch
from torchvision import transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

### Custom Dataset

```python
from CustomDataLoader import CustomDataLoader
```

```python
# transform
trans=transforms.Compose([
    transforms.ToTensor(),
    transforms.ConvertImageDtype(torch.float),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# dataset
dataset = CustomDataLoader(fmri_file = './data/fmri/sub-01_perceptionNaturalImageTraining_original_VC.h5',
                            imagenet_folder= './data/images/training',
                           transform=trans)

# dataloader
dataloader = DataLoader(dataset, batch_size=100, shuffle=True)
```

### Test Outputs

```python
for image_data, fmri_data in dataloader:
    image_data = image_data
    fmri_data = fmri_data
    break
```
